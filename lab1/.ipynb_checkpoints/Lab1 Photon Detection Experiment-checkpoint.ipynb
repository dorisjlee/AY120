### Standard deviation of the mean

```python
np.arange(0,400,10)
```

Looping through ``nstep`` so that we can change the size of the dataset chunk and compute the mean for each. Then use the mean that we found as the average so that values in ``marr`` has something to compare against.

$$ s=\sqrt{\frac{\displaystyle  \sum^N_{i=1}(\mu_i-\mu)^2}{N-1}}$$

```python
marr = np.array([]) 
i = np.arange(dt.size)
stdlist =[]
nsteplist=[]
for nstep in np.arange(0,400,10)[1:]:
    marr = np.array([]) 
    for j in i[0::nstep]: # iterate in strides of nstep
         m = np.mean(dt[j:j+nstep]) # compute the mean
         marr = np.append(marr,m) # append the new value 
    mu = np.sum(marr)/np.float(marr.size)
    std = np.sqrt(np.sum((marr - mu)**2.)/(np.float(marr.size)-1.))
    nsteplist.append(nstep)
    stdlist.append(std)
    plt.plot(nstep,std,'o')
plt.ylabel("Standard deviation of the mean [ticks]")
plt.xlabel("Number of events averaged")
```

```python
stdlist=np.array(stdlist)
nsteplist=np.array(nsteplist)
plt.ylabel("Standard deviation of the mean [ticks]")
plt.xlabel("1/sqrt(N)")
x = 1./np.sqrt(nsteplist)
y = stdlist
plt.plot(x,y,'o')
# Best fit line
par = np.polyfit(x, y, 1, full=True)
slope=par[0][0]
intercept=par[0][1]
xl = [min(x), max(x)]
yl = [slope*xi + intercept  for xi in xl]
plt.plot(xl, yl, '-g')
```

#### Adding plotting functions to analysis method

```python
def analysis(file):
    x = np.loadtxt(file,delimiter=',',dtype=np.int32)
    t = x[:,1]
    dt =t[1:] - t[0:-1]
    plt.figure()
    plt.xlabel('Event number')
    plt.ylabel('Clock tick')
    plt.plot(t) #clock cycle plot
    plt.figure()
    plt.xlabel('Event number')
    plt.ylabel('Interval [Clock ticks]')
    plt.plot(dt,',') # interval plot
    # Mean Interval Plot
    marr = np.array([]) # define an empty array
    i = np.arange(dt.size)
    nstep = 1000
    jlist=[]
    for j in i[0::nstep]: # iterate in strides of nstep
         m = np.mean(dt[j:j+nstep]) # compute the mean
         marr = np.append(marr,m) # append the new value 
         jlist.append(j)
    plt.figure()            
    plt.ylabel("Mean Interval[clock ticks]")
    plt.xlabel("Start Index")
    plt.plot(jlist,marr,'o')
```

```python
analysis("msjidlmh_140906_1057_50.csv") #truncating the ouput for display purposes
```

When ``dtype='int32`` is left out, the datalist assumes the default return type of ``loadtxt`` (floating point). Since floating point numbers are stored with limited precision in Python, it is forced to approximate the integer as a float. As a result, the last few digits of the integer may be off. Since the values of two neighboring data points we are subtracting  to obtain ``dt`` usually lie very close since they occur right after another, therefore errors introduced by the floating point approximation results in a visible difference when we are trying to compare its difference with a subsequent datapoint as shown in the plot below.

```python
x = np.loadtxt("msjidlmh_140906_1057_50.csv",delimiter=',',dtype='int32')
print x
t = x[:,1]
print t
```
