```python
import glob
def average(glob_text,title="Spectroscopic Spectra",plot = False):
    data_lst = glob.glob(glob_text)
    num_sets=len(data_lst)
    #Lets just print the first 10 for now...
    x = np.zeros((2048,),'float64')
    I = np.zeros((2048,),'float64')
    for i in data_lst:
        data = np.genfromtxt(i, skip_header=17, skip_footer=1)
        x = x + data[:,0]
        I = I + data[:,1]
    if plot:
        plt.figure()
        plt.title(title)
        plt.xlim(0,2048)
        plt.xlabel("Pixel Number",fontsize=12)
        plt.ylabel("Intensity",fontsize=12)
        plt.plot(x/num_sets,I/num_sets)
    print "Number of samples: ",num_sets
    return np.array([x,I])
```

# Finding Centroid on Mercury Spectral Lines

Looks like there are less of these, easier to use eyeball method

```python
def centroid(data):
    '''
    simple centroid calculation, assuming that the data is preprocessed
    '''
    centroid_lst=[]
    numerator = float(np.sum(data[:,1]*data[:,0]))
    cval=numerator/np.sum(data[:,1])
    return cval
```

# Wavelength calibration

Determine the wavelength calibration of the spectrometer, i.e., the mapping between
pixel number and wavelength. Do this by measuring the centroids (i.e., pixel positions) of bright Hg I (fluorescent room lights) and Ne I lines of known wavelength. Then use the method of linear least squares to determine a polynomial fit to these data.

```python
# the pix value right now are fake, used for testing only . I will need to substitute this with my centroid values when I figure that out.
pix = np.array([102, 309, 480, 1114,  1305, 1320]) # pixel position from center (from centroid calculation)
wavelength = np.array([365.0153,404.6563,435.8328,546.0735,576.9598,579.0663]) #given from NIST
assert len(pix) == len(wavelength)
z = np.polyfit(pix, wavelength, 3) #lets fit to 5th order, but we'll probably find that the  3,4th order coefficient is 0  (becareful not to overfit) 
print z
p = np.poly1d(z)
print p

fig = plt.figure()
ax = fig.add_subplot(311)
plt.plot(p(np.arange(0,10)))
plt.title("Wavelength",fontsize=12)
plt.ylabel("$\lambda$[nm]",fontsize=12)
plt.xlabel("Pixel",fontsize=12)
ax.text(0.01, 0.8,"$\lambda_i = {}x^3+{}x^2+{}x+{}$".format(z[0],z[1],z[2],z[3]), fontsize=12,transform=ax.transAxes)
```

```python
residual=(wavelength-p(pix))**2
```

```python
plt.title("Third order residual",fontsize =12)
plt.xlabel("Pixel",fontsize =12)
plt.ylabel("$\lambda_ {error}$[nm]",fontsize =12)
plt.plot(residual,'o')
```

```python
def subtract(datafile,darkfile,plot=True):
    data = np.genfromtxt(datafile, skip_header=17, skip_footer=1)
    darkdata = np.genfromtxt(darkfile, skip_header=17, skip_footer=1)
    if (plot):
        fig = plt.figure()
        fig.add_subplot(311)
        plt.xlabel("Pixel Number",fontsize=12)
        plt.ylabel("Intensity",fontsize=12)
        plt.title("Neon Spectra and Bias frames")
        plt.plot(darkdata[:,1],color="red",label="Dark Counts")
        plt.plot(data[:,1],label ="Signal")
        plt.xlim(0,2032)
        plt.ylim(0,500)
        from matplotlib.legend_handler import HandlerLine2D
        plt.legend(loc='upper left',prop={'size':10})
    print len(data)==len(darkdata)
    subtracted = data-darkdata
    if plot :
        fig2 = plt.figure()
        fig2.add_subplot(211)
        plt.plot(data[:,1])
        plt.xlim(1000,2032)
        fig2.add_subplot(212)
        plt.plot(subtracted[:,1])
        plt.xlim(1000,2032)
```
