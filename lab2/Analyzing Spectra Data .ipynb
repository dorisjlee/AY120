### Determine the wavelength calibration of the spectrometer, i.e., the mapping between pixel number and wavelength. Do this by measuring the centroids (i.e., pixel positions) of bright Hg I (fluorescent room lights) and Ne I lines of known wavelength. Then use the method of linear least squares to determine a polynomial fit to these data.

```python
sum=0

def centroid(datafile,step):
    centroid_lst=[]
    data = np.array(np.genfromtxt(datafile, skip_header=17, skip_footer=1))
    #Using element-wise operations to do a sanity check, skewed centroid since no fitting implementation
    numerator = np.sum(data[:,1]*data[:,0])
    cval_whole=numerator/np.sum(data[:,1])
    # If caclulate with this procedure and set pixel width as one I should get identical result as above 
    denominator=0
    numerator = 0
    for i in np.arange(1,max(data[:,0]),step):
#         print (data[i+1,0]-data[i,0])
        numerator+=data[i,0]*data[i,1]
        denominator += data[i,1]
        centroid_lst.append(numerator/denominator)
    return centroid_lst
```

# Smoothing : Averaging over all data sets

```python
import glob
def average(glob_text,title="Spectroscopic Spectra",plot = False):
    data_lst = glob.glob(glob_text)
    num_sets=len(data_lst)
    #Lets just print the first 10 for now...
    x = np.zeros((2048,),'float64')
    I = np.zeros((2048,),'float64')
    for i in data_lst:
        data = np.genfromtxt(i, skip_header=17, skip_footer=1)
        x = x + data[:,0]
        I = I + data[:,1]
    if plot:
        fig = plt.figure()
        ax = fig.add_subplot(211)
        plt.title(title)
        plt.xlim(0,2048)
        plt.xlabel("Pixel Number",fontsize=12)
        plt.ylabel("Intensity[ADU]",fontsize=12)
        plt.plot(x/num_sets,I/num_sets)
        for tick in ax.xaxis.get_major_ticks():
            tick.label.set_fontsize(12) 
        for tick in ax.yaxis.get_major_ticks():
            tick.label.set_fontsize(12) 
    print "Number of samples: ",num_sets
    return np.array([x,I])
```

### Boxcar smoothing

boxcar average for cosmetic purposes. lower s/N but resolution decrease. In this case could be useful for better centroid findindg 
Since the quality of each pixel is inconsistent, we employ boxcar smoothing to compensate for the 
Boxcar smoothing takes the values of adjacent pixels and average them. 
This produced undesirable results so we resorted to the conventional averaging of data by averaging over whole datasets

```python
import glob
def boxcar(datafile,interval_avrged):
#     data_lst = glob.glob(glob_text)
    #Lets just print the first 10 for now...
    source = datafile.split("_")[0]
    x = []
    I = []
    plt.figure()
    plt.title("Boxcar average of {} spectra".format(source))
    plt.xlabel("Pixel Number",fontsize=12)
    plt.ylabel("Intensity[ADU]",fontsize=12)
    plt.xlim(0,2048)
    data = np.genfromtxt(datafile, skip_header=17, skip_footer=1)
    #boxcar average
    n=0
    for j in np.arange(2048):
        if (n==interval_avrged):
            I.append(np.mean(data[j:j+interval_avrged,1]))
            x.append(j+n/2.)#center position defined as half of averaged interval 
            n=0
        else:
            n+=1
    print(len(x))
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(12) 
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(12) 
    plt.plot(x,I)
    plt.plot(x,I,'.',color='red')
```
