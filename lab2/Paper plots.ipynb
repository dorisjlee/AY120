```python
#Check the units of Intensity???
def plot_spectra(filename,title="Source Spectra"):
    data = np.genfromtxt(filename, skip_header=17, skip_footer=1)
    ax = plt.figure()
    plt.title(title)
    plt.xlabel("Pixel Number",fontsize=12)
    plt.ylabel("Intensity[ADU]",fontsize=12)
    plt.xlim(0,2048)
    plt.plot(data[:,1])
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(12) 
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(12) 
```
