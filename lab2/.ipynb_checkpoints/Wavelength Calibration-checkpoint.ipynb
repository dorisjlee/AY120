```python
#Check the units of Intensity???
def plot_spectra(filename,title="Spectroscopic Spectra"):
    data = np.genfromtxt(filename, skip_header=17, skip_footer=1)
    plt.figure()
    plt.title(title)
    plt.xlabel("Pixel Number",fontsize=12)
    plt.ylabel("Intensity",fontsize=12)
    plt.xlim(0,2048)
    plt.plot(data[:,1])
```

```python
import glob
def average(glob_text,title="Spectroscopic Spectra",plot = False):
    data_lst = glob.glob(glob_text)
    num_sets=len(data_lst)
    #Lets just print the first 10 for now...
    x = np.zeros((2048,),'float64')
    I = np.zeros((2048,),'float64')
    for i in data_lst:
        data = np.genfromtxt(i, skip_header=17, skip_footer=1)
#         x = x + data[:,0]
        x = data[:,0]
        I = I + data[:,1]
    if plot:
        plt.figure()
        plt.title("100 sample of 1s dark counts")
        plt.xlim(0,2048)
        plt.xlabel("Pixel Number",fontsize=12)
        plt.ylabel("Intensity[ADU]",fontsize=12)
        plt.plot(x,I/num_sets)
    print "Number of samples: ",num_sets
    return np.array([x,I/num_sets])
```

# Finding Centroid on Mercury Spectral Lines

Looks like there are less of these, easier to use eyeball method

```python
def centroid(data):
    '''
    simple centroid calculation, assuming that the data is preprocessed
    '''
    centroid_lst=[]
    numerator = float(np.sum(data[:,1]*data[:,0]))
    cval=numerator/np.sum(data[:,1])
    return cval
```

```python
def eyeball(start,end):
#     data  = average("roomlight_*",plot=True)
    filename="roomlight_50ms_0000.txt"
    data = np.genfromtxt(filename, skip_header=17, skip_footer=1)
    plt.plot(data[start:end,0],data[start:end,1])
    ax = plt.gca()
    cval = centroid(data[start:end])
    plt.plot([cval,cval],[ax.get_ylim()[0],ax.get_ylim()[1]],'r',linestyle='dashdot');
    print cval
    return cval
```

# Wavelength calibration

Determine the wavelength calibration of the spectrometer, i.e., the mapping between
pixel number and wavelength. Do this by measuring the centroids (i.e., pixel positions) of bright Hg I (fluorescent room lights) and Ne I lines of known wavelength. Then use the method of linear least squares to determine a polynomial fit to these data.

```python
pix_neon = np.array(cval_lst_neon)
print pix_neon
cval_lst = np.array([314.39285714285717, 483.01329243353786, 1115.5486232866895, 1374.8263590624617, 1529.818681126465]) # pixel position from center (from centroid calculation)
pix_hg = np.array(cval_lst)
print pix_hg
pix = np.concatenate([pix_neon,pix_hg])
print pix
wavelength_hg = np.array([404.6563,435.8328,546.0735,576.9598,579.0663]) #given from NIST
wavelength_neon = np.array([585.24879,594.48342,614.30626,616.35939,621.72812,626.6495,633.44278,638.29917,640.2248,650.65281,659.89529,667.82762]) #given from NIST
wavelength = np.concatenate([wavelength_neon,wavelength_hg])
print wavelength
# print len(pix)
# print len(wavelength)
assert len(pix) == len(wavelength)
z = np.polyfit(pix, wavelength, 3) #lets fit to 5th order, but we'll probably find that the  3,4th order coefficient is 0  (becareful not to overfit) 
print z
p = np.poly1d(z)
print p

fig = plt.figure()
ax = fig.add_subplot(311)
plt.plot(np.arange(0,2032),p(np.arange(0,2032)))
plt.xlim(0,2032)
plt.title("Wavelength",fontsize=12)
plt.ylabel("$\lambda$[nm]",fontsize=12)
plt.xlabel("Pixel",fontsize=12)
ax.text(0.01, 0.8,"$\lambda_i = {}x^3+{}x^2+{}x+{}$".format(z[0],z[1],z[2],z[3]), fontsize=12,transform=ax.transAxes)
```

```python
def subtract(datafile,darkfile,plot=True):
    data = np.genfromtxt(datafile, skip_header=17, skip_footer=1)
    darkdata = np.genfromtxt(darkfile, skip_header=17, skip_footer=1)
    if (plot):
        fig = plt.figure()
        fig.add_subplot(311)
        plt.xlabel("Pixel Number",fontsize=12)
        plt.ylabel("Intensity",fontsize=12)
        plt.title("Neon Spectra and Bias frames")
        plt.plot(darkdata[:,1],color="red",label="Dark Counts")
        plt.plot(data[:,1],label ="Signal")
        plt.xlim(0,2032)
        plt.ylim(0,500)
        from matplotlib.legend_handler import HandlerLine2D
        plt.legend(loc='upper left',prop={'size':10})
    print len(data)==len(darkdata)
    subtracted = data-darkdata
    if plot :
        fig2 = plt.figure()
        fig2.add_subplot(211)
        plt.plot(data[:,1])
        plt.xlim(1000,2032)
        fig2.add_subplot(212)
        plt.plot(subtracted[:,1])
        plt.xlim(1000,2032)
    return subtracted
```

```python
mulst = []
varlst=[]
for i in mega_Idata_subtracted:
    mu = np.mean(i)
    var = np.std(i)**2
    mulst.append(mu)
    varlst.append(var)
plt.ylabel("Variance[ADU$^2$]")
plt.xlabel("Mean[ADU]")
# plt.xlim(8.3*10**2,8.5*10**2)
plt.loglog(mulst,varlst,'.')
```
