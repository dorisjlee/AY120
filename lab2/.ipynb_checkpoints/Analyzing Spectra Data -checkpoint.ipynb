```python
#Check the units of Intensity???
def plot_spectra(filename,title="Spectroscopic Spectra"):
    data = np.genfromtxt(filename, skip_header=17, skip_footer=1)
    plt.figure()
    plt.title(title)
    plt.xlabel("Pixel Number",fontsize=12)
    plt.ylabel("Intensity",fontsize=12)
    plt.xlim(0,2048)
    plt.plot(data[:,1])
```

```python
plot_spectra("neon_700ms_0000.txt")
```

```python
plot_spectra("roomlight_50ms_0000.txt")
```

```python
plot_spectra("plasma_bulb0000.txt","Spectra of plasma bulb")
```

```python
plot_spectra("redcap_100ms_0000.txt")
```
