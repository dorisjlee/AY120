### Time Series of total integrated intensity on CCD

```python
intensity_lst = []
for i in glob.glob("solar-scan-11-25-9-*.fits"):
    x = pf.getdata(i)
    intensity_lst.append(sum(x))
```

```python
plt.plot(intensity_lst,'o')
```
