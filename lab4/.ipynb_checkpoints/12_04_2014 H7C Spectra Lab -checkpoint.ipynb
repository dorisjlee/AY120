```python
def fit_and_plot(x,y):
    x=1./x**2
    y=1./y
    plt.plot(x,y,'o')
    z = np.polyfit(x,y, 1) 
    print z
    p = np.poly1d(z)
    print p
#     plt.plot(p(np.arange(400,700)))
    plt.title("Calibration Curve",fontsize=12)
    plt.ylabel("$\lambda$[nm]",fontsize=12)
    return p
```

```python
fit_and_plot(helium_data,nominal_wavelength)
print z[1]
print p[1]
```
