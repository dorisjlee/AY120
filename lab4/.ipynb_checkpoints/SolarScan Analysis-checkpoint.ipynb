### Intensity relation obtained from Eddington approximation (C&O Eq. 9-58)

$\frac{I(\theta)}{I(\theta=0)}= \frac{2}{5}+\frac{3}{5}cos\theta$

let b = impact parameter be your varying parameter so you know how far across you are cutting across the sun

$cos\theta = \frac{\sqrt{R^2-b^2}}{R}$

```python
import numpy as np
from scipy.optimize import curve_fit
def func(x, a, b, c):
    return a * np.exp(-b * x) + c
xdata = np.linspace(0, 4, 50)
y = func(xdata, 2.5, 1.3, 0.5)
ydata = y + 0.2 * np.random.normal(size=len(xdata))
popt, pcov = curve_fit(func, xdata, ydata)
plt.plot(xdata,ydata,'o')
plt.plot(xdata,func(xdata,popt[0],popt[1],popt[2]))
```
