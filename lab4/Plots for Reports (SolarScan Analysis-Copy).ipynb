```python
#Computing Covariance
#Let x = start
#Let y = end
spectra_1d_start = np.array(spectra_1d_start)/median(spectra_1d_start)
spectra_1d_end = np.array(spectra_1d_end)/median(spectra_1d_end)
xbar = mean(spectra_1d_start)
ybar = mean(spectra_1d_end)
N=len(spectra_1d_start)
s_xy = 1./(N-1)*sum((spectra_1d_start-xbar)*(spectra_1d_end-ybar))
```

#### Auto-covariance

$y_i = x_{i+j}$

```python
autocorrval= []
for i in np.linspace(-10,10,100):
    lag = i#placeholder value (need to determine this value somehow )
    xi = start
    xij = np.roll(start,int(lag))
    N=len(start)
    # #Covariance s_xy^2
    # s_xy = 1./(N-1)*sum((spectra_1d_start-xbar)*(spectra_1d_end-ybar))
    # print "Covariance (s_{xy}^2): %f" %s_xy**2
    #autocovariance
    ac = 1./(N-1)*sum(xi*xij)-N/float(N-1)*mean(xi)**2
#     print "Autocorrelation: %f" %ac
    autocorrval.append(ac)
    # # cross correlation
    # yi=spectra_1d_end
    # yij = np.roll(spectra_1d_end,n)
    # cc = 1./(N-1)*sum(xi*yij)-N/float(N-1)*mean(xi)*mean(yi)
    # print "Cross Correlation: %f" %cc
```
