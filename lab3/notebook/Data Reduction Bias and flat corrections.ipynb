The n_overscan is actually not necessary (Args from the lab handout)

```python
def bsub(data,n_overscan=32):
    offset_lst = []
    for i in x[:,x.shape[0]:]:
        offset_lst.append(mean(i))
#     print offset_lst
#     offset_lst = np.array(offset_lst)
    offset_val = mean(offset_lst)
    print offset_val
#     print offset_lst
    return data - offset_val
```

# Choosing different color tables 

```python
def display_colormaps(data):
    plt.figure()
    plt.subplot(221)
    imp = plt.imshow(data,cmap='spectral')
    plt.colorbar()
    plt.subplot(222)
    imp = plt.imshow(data,cmap='hot')
    plt.colorbar()
    plt.tight_layout()
    plt.figure()
    plt.subplot(221)
    imp = plt.imshow(data,cmap='summer')
    plt.colorbar()
    plt.subplot(222)
    imp = plt.imshow(data,cmap='afmhot')
    plt.colorbar()
    plt.tight_layout()
```

```python
display_colormaps(flatb[::-1])
```

```python
data =(xb/flatb)[::-1]
plt.figure()
plt.subplot(221)
imp = plt.imshow(data,cmap='spectral',norm=LogNorm())
plt.colorbar()
plt.subplot(222)
imp = plt.imshow(data,cmap='hot',norm=LogNorm())
plt.colorbar()
plt.tight_layout()
plt.figure()
plt.subplot(221)
imp = plt.imshow(data,cmap='summer',norm=LogNorm())
plt.colorbar()
plt.subplot(222)
imp = plt.imshow(data,cmap='gray_r',norm=LogNorm())
plt.colorbar()
plt.tight_layout()
```
